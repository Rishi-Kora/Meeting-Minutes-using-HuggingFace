# Using Whisper OpenAI to convert Audio to Text

```python
import openai
openAIsecrets = UserSecretsClient()
openai_key = openAIsecrets.get_secret("OPENAI_API_KEY")
openai.api_key = openai_key
```

```python
audio_file = open(audio_filename, "rb")
transcription = openai.audio.transcriptions.create(model=AUDIO_MODEL, file=audio_file, response_format="text")
print(transcription)
```

```python
system_message = "You are an assistant that produces minutes of meetings from transcripts, with summary, key discussion points, takeaways and action items with owners, in markdown."
user_prompt = f"Below is an extract transcript of a Denver council meeting. Please write minutes in markdown, including a summary with attendees, location and date; discussion points; takeaways; and action items with owners.\n{transcription}"
prompts = system_message+user_prompt
```

# `device_map="auto"` lets the model automatically choose the best device (CPU or GPU) for loading. 
# `load_in_4bit=True` loads the model in 4-bit precision to reduce memory usage and improve speed.

```python
from transformers import AutoTokenizer, AutoModelForCausalLM
import torch
device = "cuda" if torch.cuda.is_available() else "cpu"
tokenizer = AutoTokenizer.from_pretrained("korarishi/rishi-2-2b-it")

# 3) load the quantized model
model = AutoModelForCausalLM.from_pretrained(
    "korarishi/rishi-2-2b-it",
    device_map="auto",        
    load_in_4bit=True,        
    torch_dtype=torch.float16 
)
model.eval()
```

# Prints the Minutes of Meeting in a Markdown Format

```python
inputs = tokenizer(prompts, return_tensors="pt").to(model.device)
outputs = model.generate(
    **inputs,
    max_new_tokens=100,
    do_sample=True,
    top_p=0.95,
    temperature=0.8
)
transcription = tokenizer.decode(outputs[0], skip_special_tokens=True)
display(Markdown(transcription))
```
